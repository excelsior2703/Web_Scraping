```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.scrapethissite.com/pages/forms/"
```

```python
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
class="table"
```

```python
table=soup.find('table',class_="table")
```

```python
print(table)
```

```python
head=table.find_all('th')
```

```python
head_table=[data.text.strip() for data in head ]
```

```python
print(head_table)
```

```python
import pandas as pd

```

```python
df = pd.DataFrame(columns = head_table)
df
```

```python
column_data=soup.find_all('tr',class_='team')
print(column_data)
```

```python
for row in column_data[1:]:
    row_data=row.find_all('td')
    ind_row=[data.text.strip() for data in row_data]
  
    length = len(df)
    df.loc[length]=ind_row
    
```
